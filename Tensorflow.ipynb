# Tensorflow Math **Modules**

```python
import tensorflow as tf
import numpy as np
```

```python
a=10
b=20
```

```python
a+b
```

```python
tf.add(a,b)
```

```python
# addition two list
l1=[1,2,3,4]
l2=[4,5,6,7]
tf.add(l1,l2)
```

```python
# addition two tuple
t1=(2,3,4,5)
t2=(4,6,7,8)
tf.add(t1,t2)
```

```python
l=[10,21,31,14]
a=23
tf.add(l,a)
```

```python
# if we find x*y
x=tf.constant([[2,3],[3,3]])
y=tf.constant([[8,12],[4,5]])
tf.pow(x,y)
```

```python
# tf.math.abs
# real number
x=tf.constant([-2.25,3.25])
tf.abs(x)
```

```python
# complex number
x=tf.constant([[-2.25 + 4.75j],[-3.25 + 5.75j]])
tf.abs(x)
```

```python
# substraction
x=[1,2,3,4,5]
y=2
tf.subtract(x,y)
```

```python
x=[1,2,3,4,5]
y=2
tf.subtract(y,x)
```

```python
# tf.matrh.multiply
x=tf.constant(([1,2,3,4]))
tf.math.multiply(x,x)
```

```python
tf.math.multiply(7,8)
```

```python
x=tf.ones([1,2]);
y=tf.ones([2,1]);
x*y
```

```python
# tf.math.maximum
x=tf.constant([0.,0.,0.,0.])
y=tf.constant([-2.,0.,2.,5.])
tf.math.maximum(x,y)
```

```python
x=tf.constant([-5.,0.,0.,0.])
y=tf.constant([-3.])
tf.math.maximum(x,y)
```

```python
# tf.math.log
x=tf.constant([0,0.5,1,5])
tf.math.log(x)
```

```python
# tf.math.divide
x=tf.constant([16,12,11])
y=tf.constant([4,6,2])
tf.math.divide(x,y)
```

```python
# tf.math.sqrt
x=tf.constant([[4.0],[16.0]])
tf.sqrt(x)
```

```python
x=tf.constant([[-4.0],[16.0]])
tf.sqrt(x)
```
