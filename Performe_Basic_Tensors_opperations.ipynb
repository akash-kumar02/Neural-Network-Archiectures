 Write a Python program to create a tensor of shape (3, 3) filled with random numbers between 0
and 1. Perform basic tensor operations including matrix addition, multiplication, and transpose.
Convert the resulting tensor into a NumPy array and then reshape it into a (1, 9) tensor.

```python
import torch
import numpy as np
```

```python
# 1: Create a 3x3 tensor with random numbers between 0 and 1.
tensor_a=torch.rand(3,3)
tensor_a
```

```python
# 2: Create another 3x3 random tensor for operations.
tensor_b=torch.rand(3,3)
tensor_b
```

```python
# 3: Perform matrix addition.
addition=tensor_a+tensor_b
addition
```

```python
# 4: Perform matrix multiplication.
multiplication=torch.matmul(tensor_a,tensor_b)
multiplication
```

```python
# OR
torch.matmul(torch.rand(3,2),torch.rand(2,3))
```

```python
# 5: Transpose the multiplication result.
Transposed=multiplication.T
Transposed
```

```python
print(type(Transposed))
```

```python
# 6: Convert to NumPy array.
numpy_array = Transposed.numpy()
numpy_array
```

```python
print(type(numpy_array))
```

```python
# 7: Reshape into (1, 9).
reshaped_tensor = torch.tensor(numpy_array.reshape(1, 9))
reshaped_tensor
```
